- Cartooning an image turns a normal photo into a fun, animated-style picture. With OpenCV, we do this by smoothing the image to simplify colors and detecting edges to create outlines. Combining these steps makes the photo look like a cartoon.

```python
import cv2

file = "image.png"

img = cv2.imread(file)
if img is None:
    print("Image not found")
    exit()

g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
g = cv2.medianBlur(g, 5)

e = cv2.adaptiveThreshold(
    g, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY, 9, 9
)

c = cv2.bilateralFilter(img, 9, 250, 250)

cartoon = cv2.bitwise_and(c, c, mask=e)

cv2.imshow("Cartoon", cartoon)
cv2.imwrite(file, cartoon)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
list_img = list([img, cartoon])

for i in range(2):

    plt.subplot(2, 1, i+1)
    image = list_img[i]

    plt.imshow(image)
    plt.title(f'Sample cartoon image')
    plt.subplot
    plt.show()
```
